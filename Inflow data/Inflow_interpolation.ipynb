```python
import numpy as np
import matplotlib.pyplot as plt

# Define the filename
filename = '/content/drive/MyDrive/ColabNotebooks/Tesi/Inflow_profiles/R_ICA.csv'

# Read the CSV file using numpy
data = np.loadtxt(filename, delimiter=',', skiprows=1)  # Adjust 'skiprows' if there's no header

# Extract the timing and velocity columns
timings = data[:, 0]  # First column: timings
velocities = data[:, 1]  # Second column: velocities

# Define new timings using linspace
T = 1
dt = 1e-5
num_points = int(np.ceil(T / dt))
new_timings = np.linspace(timings.min(), timings.max(), num_points + 1)  # Adjust the number of points as needed

# Interpolate velocities on the new timings
new_velocities = np.interp(new_timings, timings, velocities)

# Plot the original and interpolated data
plt.figure()
plt.plot(timings, velocities, 'o', label='Original Data')  # Original data points
plt.plot(new_timings, new_velocities, '-', label='Interpolated Data')  # Interpolated curve
plt.title('Velocity Interpolation')
plt.xlabel('Time')
plt.ylabel('Velocity')
plt.legend()
plt.grid(True)
plt.show()

# Optionally, save the interpolated data to a new CSV file
interpolated_data = np.column_stack((new_timings, new_velocities))
np.savetxt('/content/drive/MyDrive/ColabNotebooks/Tesi/Inflow_profiles/R_ICA_inflow.csv', interpolated_data, delimiter=',', header='Time,Velocity', comments='')
```
